```python
#Importing-Libraries#

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
#----------------------------------------------------------
```

```python
tr_data = pd.read_csv('...')
tr_data.drop('Id', axis=1, inplace=True)

tt_data = pd.read_csv('...')
tt_data.drop('Id', axis=1, inplace=True)

feature = pd.read_csv('...')
feature.drop('Id', axis=1, inplace=True)
```

```python
# The colums who has 45% missing values dropping

Tr_Ms_Values = tr_data.isnull().sum()
Tr_Percent = (45/100) * len(tr_data)

print('Drop the columns which has 45% missing values')
print()
print()

for i in tr_data:
    #print('The columns: {} has {:.2f}% missing values'.format(i,(Ms_Values[i]/len(Tr_Data)*100)))
    if Tr_Ms_Values[i] > Tr_Percent:
        print()
        print('The columns: {} has {:.2f}% missing value and will be deleted'.format(i,(Tr_Ms_Values[i]/len(tr_data)*100)))
        print()
        tr_data.drop(i,axis=1, inplace=True)
        tt_data.drop(i,axis=1, inplace=True)
```

```python
tr_data.info()
```

```python
for i in tr_data:
    
    if (tr_data[i].dtype == 'object') and (tr_data[i].isnull().sum() > 0 ):
        print('The columns is {} and have {} null values from {} and data-type is object.'.format(i,tr_data[i].isnull().sum(), len(tr_data.index)))
        print(tr_data[i].value_counts()) 
        print()
    elif ((tr_data[i].dtype == 'int') or (tr_data[i].dtype == 'float')) and (tr_data[i].isnull().sum() > 0 ):
        print('The columns is {} and have {} null values from {} and data-type is {}.'.format(i,tr_data[i].isnull().sum(),len(tr_data.index),tr_data[i].dtype))
        tr_data[i].bfill(axis ='rows', inplace=True) # Fill across the row
        print()
        
        
tr_data['MasVnrType'].fillna('None', inplace=True)
tr_data['BsmtQual'].fillna('TA', inplace=True)
tr_data['BsmtCond'].fillna('TA', inplace=True)
tr_data['BsmtExposure'].fillna('No', inplace=True)
tr_data['BsmtFinType1'].fillna('Unf', inplace=True)
tr_data['BsmtFinType2'].fillna('Unf', inplace=True)
tr_data['Electrical'].fillna('SBrkr', inplace=True)
tr_data['GarageType'].fillna('Attchd', inplace=True)
tr_data['GarageFinish'].fillna('Unf', inplace=True)
tr_data['GarageQual'].fillna('TA', inplace=True)
tr_data['GarageCond'].fillna('TA', inplace=True)
```

```python
for i in tt_data:
    
    if (tt_data[i].dtype == 'object') and (tt_data[i].isnull().sum() > 0 ):
        print('The columns is {} and have {} null values from {} and data-type is object.'.format(i,tt_data[i].isnull().sum(), len(tt_data.index)))
        print(tt_data[i].value_counts()) 
        print()
    elif ((tt_data[i].dtype == 'int') or (tt_data[i].dtype == 'float')) and (tt_data[i].isnull().sum() > 0 ):
        print('The columns is {} and have {} null values from {} and data-type is {}.'.format(i,tt_data[i].isnull().sum(),len(tt_data.index),tt_data[i].dtype))
        tt_data[i].bfill(axis ='rows', inplace=True) # Fill across the row
        print()
        
        
tt_data['MSZoning'].fillna('RL', inplace=True)
tt_data['Utilities'].fillna('AllPub', inplace=True)
tt_data['Exterior1st'].fillna('VinylSd', inplace=True)
tt_data['Exterior2nd'].fillna('VinylSd', inplace=True)
tt_data['MasVnrType'].fillna('None', inplace=True)
tt_data['BsmtQual'].fillna('TA', inplace=True)
tt_data['BsmtCond'].fillna('TA', inplace=True)
tt_data['BsmtExposure'].fillna('No', inplace=True)
tt_data['BsmtFinType1'].fillna('GLQ', inplace=True)
tt_data['BsmtFinType2'].fillna('Unf', inplace=True)
tt_data['KitchenQual'].fillna('TA', inplace=True)
tt_data['Functional'].fillna('Typ', inplace=True)
tt_data['GarageType'].fillna('Attchd', inplace=True)
tt_data['GarageFinish'].fillna('Unf', inplace=True)
tt_data['GarageQual'].fillna('TA', inplace=True)
tt_data['GarageCond'].fillna('TA', inplace=True)
tt_data['SaleType'].fillna('WD', inplace=True)
```

```python
#Encode the categorical data in train and test dataframe
for i in tr_data:
    if(tr_data[i].dtype == 'object'):
        tr_data[i] = labelencoder.fit_transform(tr_data[i])
        
for i in tt_data:
    if(tt_data[i].dtype == 'object'):
        tt_data[i] = labelencoder.fit_transform(tt_data[i])
```

```python
X = tr_data.iloc[:, 0:74]
Y = tr_data.iloc[:, 74]

#Split the Train-Data to 80%-Train and 20-Test
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
lr.fit(x_train, y_train)
print('Linear-Regression Accuracy score is: ' , lr.score(x_train,y_train))

# Make predictions using the testing set
y_pred = lr.predict(x_test)
```
